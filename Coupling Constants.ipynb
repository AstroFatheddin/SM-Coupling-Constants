plotting $\alpha^{-1}$ (Coupling Constants) for Electromagentic, Weak and Strong Force

```python
import numpy as np
import math as mat
import matplotlib.pyplot as plt
%matplotlib inline

mz=91.1876 #[GeV]
def a1(mu):
    return (59-2.3*41/(20*np.pi) *(mu-np.log(mz)))
def a2(mu):
    return (30+2.3*19/(12*np.pi) *(mu-np.log(mz)))
def a3(mu):
    return (8.5+2.3*7/(2*np.pi) *(mu-np.log(mz)))

mu=2
M=[]
while mu<=22:
    mu=mu+0.1
    M.append(mu)
    
A1=[]
A2=[]
A3=[]
for i in range (len(M)):
    A1.append(a1(M[i]))
    A2.append(a2(M[i]))
    A3.append(a3(M[i]))
    
plt.plot(M,A1,label='EM Force')
plt.plot(M,A2,label='Weak Force')
plt.plot(M,A3, label='Strong Force')
plt.grid(True, which='both')
plt.xlabel('Log(GeV)')
plt.ylabel('a**(-1)')
plt.legend()
plt.show()
```
